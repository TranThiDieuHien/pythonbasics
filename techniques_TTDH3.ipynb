```
from bs4 import BeautifulSoup
import requests
import os
```

1 Tải 1 url về

```
url = 'https://vietnamnet.vn/vn/kinh-doanh/dau-tu/nha-dau-tu-eu-muon-rot-ty-usd-xay-trung-tam-logistics-o-phu-my-704586.html'
#Gửi yêu cầu truy cập url
page = requests.get(url)
soup = BeautifulSoup(page.text, "html.parser")
html = soup.prettify()
print(html)
```

2. Lưu 1 html vào thư mục chỉ định

```
path = 'C:\\Users\\DELL\\'
os.chdir(path)
t = 'url_list'
os.mkdir(t) 
path1= path + t
os.chdir(path1)
file = "html_file" + ".txt"
f = open(file, "w")
f.write(html)
f.close()
print(f)
```

4. Cách lấy các url trong html

```
url_list=[]
raw = soup.find_all("a")
print(len(raw))
for item in raw :
   link = item.get("href")
   #print(link)
   url_list.append(link)
#print(len(url_list))
for item in url_list:
    item = str(item)
    if not(item.find("http", 0, 4)) :
        print(item)           
                      

```
